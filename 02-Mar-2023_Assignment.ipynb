# Q1 What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a Python library used for creating static, animated, and interactive visualizations in Python. It is a powerful tool for data visualization that allows users to create high-quality charts, plots, histograms, and other visualizations in Python.

Matplotlib is used for a wide variety of applications, including scientific research, data analysis, and machine learning. It is particularly useful for exploring and analyzing data, and for presenting results in a clear and concise manner.

Five plots that can be plotted using the Pyplot module of Matplotlib are:

    Line Plot: A line plot displays data as a series of data points connected by straight lines. It is commonly used to visualize trends over time or to show how one variable changes in response to another.

    Scatter Plot: A scatter plot displays data as a collection of points on a 2D plane. It is commonly used to visualize the relationship between two variables.

    Bar Plot: A bar plot displays data as rectangular bars of different heights or lengths. It is commonly used to compare the values of different variables or to show how one variable changes over time.

    Histogram: A histogram displays the distribution of a continuous variable by dividing the data into a set of bins and counting the number of data points that fall into each bin.

    Pie Chart: A pie chart displays data as a circular graph, with each segment representing a different category or value. It is commonly used to show the proportion of each category within a larger dataset.



# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

A scatter plot is a type of plot that displays the relationship between two variables as a collection of points on a 2D plane. Each point on the plot represents a single data point, and the position of the point is determined by the values of the two variables being plotted.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(3)
```

```python
x = 3 + np.random.normal(0,2, 50)
y = 3 + np.random.normal(0,2, len(x))
```

```python
plt.scatter(x,y)
plt.title("scatter Plot of X vs Y")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows us to plot multiple graphs side by side or stacked on top of each other, making it easier to compare different plots or visualize multiple aspects of a dataset at once.

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0,1,2,3,4,5])
y1 = np.array([0,100,200,300,400,500,])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0,1,2,3,4,5])
y3 = np.array([10,20,30,40,50,60])

x4 = np.array([0,1,2,3,4,5])
y4 = np.array([200,350,250,550,450,150])
```

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)

axs[0, 0].set_title('Line 1')
axs[0, 1].set_title('Line 2')
axs[1, 0].set_title('Line 3')
axs[1, 1].set_title('Line 4')

axs[1, 0].set_xlabel('X')
axs[1, 1].set_xlabel('X')
axs[0, 0].set_ylabel('Y')
axs[1, 0].set_ylabel('Y')
plt.tight_layout()
plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

A bar plot is a type of chart that displays categorical data with rectangular bars, where the height or length of each bar represents the value of the data being plotted. It is commonly used to show comparisons between different categories of data or to track changes over time.

A horizontal bar plot is a type of bar plot where the bars are plotted horizontally instead of vertically. It is useful when the category labels are long or the values being plotted have long labels.

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company, profit)
plt.title('Company Profits')
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.show()
```

```python
plt.barh(company, profit)
plt.title('Company Profits')
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.show()
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset that shows the distribution of the data along a number line. It is commonly used to summarize the spread and shape of a dataset, as well as to identify potential outliers.

The box in a box plot represents the interquartile range (IQR), which contains the middle 50% of the data. The whiskers extend from the box to show the range of the data, excluding potential outliers. Outliers are individual data points that are more than 1.5 times the IQR away from the box.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]

plt.boxplot(data)
plt.title('this is my box plot')
plt.xticks([1, 2], ['box 1' , 'box 2'])
plt.ylabel('values')
plt.show()
```
