# Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

```python
import seaborn as sns
import plotly.express as px
```

```python
titanic_df = sns.load_dataset('titanic')


fig = px.scatter(titanic_df, x='age', y='fare')
fig.show()
```

# Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
tips_df = px.data.tips()
fig = px.box(tips_df, x='day', y='total_bill', color='sex')
fig.show()
```

# Multiple box

```python
fig = make_subplots(rows=1, cols=4, subplot_titles=('Thur', 'Fri', 'Sat', 'Sun'))

for i, day in enumerate(['Thur', 'Fri', 'Sat', 'Sun']):
    df = tips_df[tips_df['day'] == day]
    fig.add_trace(go.Box(y=df['total_bill'], name=day), row=1, col=i+1)
    
fig.update_layout(title='Box plots of total bill by day', xaxis_title='Day', yaxis_title='Total bill')
fig.show()
```

# Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

```python
import plotly.express as px

tips_df = px.data.tips()

fig = px.histogram(tips_df, x="sex", y="total_bill", color="day", pattern_shape="smoker")
fig.show()
```

# Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.

Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
dimensions parameter.

```python
iris_df = px.data.iris()

fig = px.scatter_matrix(iris_df, dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"], color="species")
fig.show()
```

# Q5. What is Distplot? Using Plotly express, plot a distplot.

Distplot is a visualization in Seaborn library that displays a univariate distribution of observations in a dataset. It combines a histogram of the distribution with a kernel density estimate, providing a more comprehensive understanding of the distribution of a variable.

```python
import plotly.express as px

tips_df = px.data.tips()
fig = px.histogram(tips_df, x="total_bill", nbins=30, marginal="rug", opacity=0.7)
fig.show()
```

# gapminder dataset.

```python
import plotly.express as px

gapminder_df = px.data.gapminder()

fig = px.histogram(gapminder_df, x="lifeExp", color="continent", nbins=30, opacity=0.7,
                   marginal="rug", barmode="overlay", histnorm="probability density")

fig.update_layout(title="Distribution of Life Expectancy by Continent")
fig.show()
```
