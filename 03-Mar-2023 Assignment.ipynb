# 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

1.Scatter Plot: It is used to display the relationship between two variables in a dataset. It helps in identifying any patterns, trends, or clusters that may exist between the variables.

2.Line Plot: It is used to show the trend of a variable over time or some other continuous variable. It helps in identifying the changes in the variable over time.

3.Histogram: It is used to visualize the distribution of a single numerical variable. It helps in identifying the range and frequency of values for that variable.

4.Bar Plot: It is used to display and compare the values of a categorical variable. It helps in identifying the frequency or count of each category.

5.Heatmap: It is used to visualize the correlation between multiple variables in a dataset. It helps in identifying the strength and direction of the relationships between the variables.

# 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions. 
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
```

```python
fmri_data = sns.load_dataset("fmri")
```

```python
fmri_data
```

```python
sns.lineplot(x = 'timepoint' , y = 'signal' , hue = 'event' , style = 'region' , data=fmri_data)
```

# 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
titanic_data = sns.load_dataset("titanic")
```

```python
titanic_data
```

```python
sns.boxplot(x = 'pclass' , y = 'age' , data=titanic_data)
sns.boxplot(x = 'pclass' , y = 'fare' , data=titanic_data)
```

# 4.Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
diamonds_data = sns.load_dataset('diamonds')
```

```python
diamonds_data
```

```python
sns.histplot(data=diamonds_data , x = 'price' , hue = 'cut' , kde=True)
```

# 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
iris_data = sns.load_dataset('iris')
```

```python
iris_data
```

```python
sns.pairplot(data=iris_data , hue = 'species')
```

# 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights_data = sns.load_dataset('flights')
```
